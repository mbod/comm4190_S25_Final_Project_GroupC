### Overview ###
The objective of HealthPlan Helper is to help people navigate the complex US healthcare system. 

This is what the full HealthPlan Helper experience looks like for a user whose main question is "Which healthcare plan should I choose?": 

#### Steps ####

  1. Introduce yourself
  2. Ask "get to know you" questions: age, health conditions, healthcare usage over previous year, if expect to use more or less next year
  3. Ask for health insurance plan brochures and summaries
  4. Perform analysis
  5. Ask if user has any questions about the plans 
     
However, HealthPlan Helper can also help users only wanting to go through some of the steps. For example, it could still help a user predict their healthcare usage for the next year even if the user does not want to share health insurance plan brochures. It can also answer questions about how the US healthcare system works in general. See the following notebook for Mermaid diagrams of a few user scenarios. 

#### Capabilities #### 
HealthPlan Helper can: 

1. Use Retrieval-Augmented Generation to answer questions about healthcare plans.
2. Analyze and graph data to make plan recommendations.
3. Use DuckDuckGo to answer questions about the US health insurance system.
4. Adapt to the tone of the user. 
5. Stay on topic!
